# Fruits and Places

```python
import pandas as pd  # imported pandas
```

#### This csv file contains Location of the fruit farms and Fruit Names

read the content from the csv file using read_csv and saved to a new dataframe

```python
fr_loc=pd.read_csv('fruitandlocation.csv')
```

```python
fr_loc.head() # five records in the dataset
```

#### In the fruit columns there different types of fruit related to the same location. by using the wrangling techniques all the fruit names were splitted with respect to their locationa and saved to the new data frame

```python
reshaped = \
(fr_loc.set_index(fr_loc.columns.drop('Fruits',1).tolist()) # the fruit column is converted to a list by dropping it
   .Fruits.str.split(',', expand=True) # the  fruit names were divided with the delimeter ","
   .stack()
   .reset_index()
   .rename(columns={0:'Fruits'})
   .loc[:, fr_loc.columns]
)

reshaped.to_csv('fr_loc.csv', index=False)
```

```python
fr_loc1=pd.read_csv('fr_loc.csv')
fr_loc1.head()
```

#### There one more data set which has the details of the gost plants names where the fruit fly are mostly attracted to

```python
host=pd.read_csv('hostplants.csv')
```

```python
host.head()
```

```python
host=host.rename(columns={"Host Plants": "Fruits"}) #renaming the columns name to merge with the fruit and location dataset
host.head()
```

```python
from functools import reduce
host_df = [host,fr_loc1]
host_fruit_merge = reduce(lambda  left,right: pd.merge(left,right,on=['Fruits'],
                                            how='right'), host_df)  # merging the two datasets with respect to the fruit name
```

```python
host_fruit_merge=host_fruit_merge.drop(['Susceptability'], axis=1) # dropping the unwanted column
host_fruit_merge.head()
```

```python
host_fruit_merge.to_csv('fruitloc.csv', index=False)
```

# PH levels in soil and Fruit


```python
ph=pd.read_csv("phvalues.csv")
ph.head()
```

```python
ph=ph.rename(columns={"PH value fruit": "PH_value_fruit"})
ph=ph.rename(columns={"PH value soil": "PH_value_soil"})
```

```python
ph[['Ph_Levels_Fruits_LB','Ph_Levels_Fruits_UB']] = ph.PH_value_fruit.str.split("-",expand=True) # splitting the column data with ethe symbol "-"
ph[['Ph_Levels_Soil_LB','Ph_Levels_Soil_UB']] = ph.PH_value_soil.str.split("-",expand=True)
ph=ph.drop(['PH_value_fruit', 'PH_value_soil'], axis=1)
```

```python
ph.head()
```

```python
ph.to_csv('phlevels_soil_fruit.csv', index=False)
```

# Merging all the data files

```python
lat_lon=pd.read_csv("location.csv")
```

```python
lat_lon=lat_lon.rename(columns={"L_Name": "Location"})
lat_lon.head()
```

```python
fruit_loc1=pd.read_csv("fruitslocation.csv")
```

```python
fruit_loc1=fruit_loc1.rename(columns={"L_Name": "Location"})
fruit_loc1.head()
```

```python
fruit_soil_ph=pd.read_csv("phlevels_soil_fruit.csv")
```

```python
fruit_soil_ph=fruit_soil_ph.rename(columns={"Names": "Fruit"})
fruit_soil_ph.head()
```

```python
preventive=pd.read_csv("preventive_methods.csv")
```

```python
preventive.head()
```

```python
control=pd.read_csv("Control_methods.csv")
```

```python
control.head()
```

```python
seasons=pd.read_csv("Seasons.csv")
```

```python
seasons=seasons.rename(columns={"F_Name": "Fruit"})
seasons.head()
```

#### merging all the cleaned data files into one d

```python
from functools import reduce
data_frames = [seasons,control,preventive,fruit_soil_ph,fruit_loc1]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Fruit'],
                                            how='outer'), data_frames)
```

```python
data_frames1=[df_merged,lat_lon]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Location'],
                                            how='outer'), data_frames1)
```
